# Load Data

```python
path = kagglehub.dataset_download("amar5693/student-performance-dataset")
print("Path to dataset files:", path)

csvPath = os.path.join(path, "ultimate_student_productivity_dataset_5000.csv")
df = pd.read_csv(csvPath)
```

# Distribution of self study hours

```python
plt.figure(figsize=(8,5))
sns.histplot(df['self_study_hours'], bins=20)
plt.title("Distribution of Self Study Hours")
plt.show()
```

```python
academic_sleep = df.groupby('academic_level')['sleep_hours'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
academic_sleep.plot(kind='bar')
plt.title("Academic Level")
plt.ylabel("Sleep Hours")
plt.show()

academic_sleep
```

# Highest Exam Score

```python
examScore= df.groupby('student_id')['exam_score'].sum().sort_values(ascending=False)

examScore.head(10)
```

# Scatter Plot

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='focus_index', y='burnout_level', data=df)
plt.title("Focus Index vs Burnout Level")
plt.show()
```

# Correlation Matrix

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['study_hours', 'productivity_score']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
